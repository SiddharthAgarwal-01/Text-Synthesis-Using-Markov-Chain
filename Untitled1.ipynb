```python
def rem_special(col):
    text=col[0]
    text=re.sub(r'[^A-Za-z0-9]+'," ",text)
    return text
df1['PlayerLine']=df1[['PlayerLine']].apply(rem_special,axis=1)
```

```python
df1
```

```python
def rem_dig(col):
    text=col[0]
    text=re.sub(r'\d+',"",text)
    return text
df1['PlayerLine']=df1[['PlayerLine']].apply(rem_dig,axis=1)
```

```python
def con_tolower(col):
    text=col[0]
    text=text.lower()
    return text
df1['PlayerLine']=df1[['PlayerLine']].apply(con_tolower,axis=1)
```

```python
dic = {}
def fn(col):
    corpus = col[0]
    #print(corpus)
    for i in range(len(corpus)-1):
        #print(corpus[i]+' '+ corpus[i+1])
        if corpus[i] in dic.keys():
            dic[corpus[i]].append(corpus[i+1])
        else:
            dic[corpus[i]] = [corpus[i+1]]
```

```python
df1[['PlayerLine']].apply(fn,axis=1)
```
